```python
from llm.tokenization.gpt_dataset_v1 import create_dataloader_v1

with open("../the-verdict.txt", "r", encoding="utf-8") as file:
    text = file.read()

dataloader = create_dataloader_v1(
    text, batch_size=8, max_length=4, stride=4, shuffle=False
)

data_iter = iter(dataloader)
inputs, targets = next(data_iter)
print("Inputs:\n", inputs)
print("Targets:\n", targets)

print("batch_size:", inputs.shape[0])  # expected: 8
print("seq_length:", inputs.shape[1])  # expected: 4
print("targets_shape:", targets.shape)  # expected: same as inputs
print("first_two_sequences:\n", inputs[:2])
```
